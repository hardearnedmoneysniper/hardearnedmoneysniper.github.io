## 1. 先学会输出：`print()`

`print()` 的作用是：把内容显示出来。

```python
print('Hello, world!')
print('你好，Python')
```

你可以把引号里的文字改成任何你想显示的内容。

## 2. 变量：给数据起名字

变量就像一个盒子，你可以把数据放进去，然后给这个盒子起个名字。

```python
name = 'Wuyang'
age = 26
print(name)
print(age)
```

这里：
- `name` 是变量名
- `'Wuyang'` 是字符串
- `age` 是变量名
- `26` 是整数

## 3. 常见数据类型

Python 最常见的几种类型：
- `int`：整数
- `float`：小数
- `str`：字符串
- `bool`：真假值（True / False）

```python
a = 10
b = 3.14
c = 'python'
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

## 4. 最基础运算

Python 可以像计算器一样做加减乘除。

```python
x = 10
y = 3

print(x + y)   # 加法
print(x - y)   # 减法
print(x * y)   # 乘法
print(x / y)   # 除法
print(x ** y)  # 次方
```

## 5. 字符串拼接

字符串就是文字。你可以把文字和变量拼起来。

```python
name = 'Wuyang'
school = 'Westlake University'

print('My name is ' + name)
print(f'I study at {school}.')
```

这里第二种写法 `f'...'` 很常用，叫 **f-string**。
以后你会经常用它。

## 6. 列表 `list`

列表可以把多个东西放在一起。

```python
fruits = ['apple', 'banana', 'orange']
print(fruits)
print(fruits[0])
print(fruits[1])
```

注意：Python 下标从 `0` 开始。

所以：
- `fruits[0]` 是第 1 个元素
- `fruits[1]` 是第 2 个元素

## 7. `if` 判断

`if` 用来根据条件决定做什么。

```python
score = 85

if score >= 60:
    print('Pass')
else:
    print('Fail')
```

注意 Python 的缩进非常重要。
`print('Pass')` 前面的 4 个空格不能乱。

## 8. `for` 循环

循环就是把一件事情重复做很多次。

```python
for i in range(5):
    print(i)
```

`range(5)` 会产生：`0, 1, 2, 3, 4`。

```python
fruits = ['apple', 'banana', 'orange']

for fruit in fruits:
    print(fruit)
```

## 9. 函数 `def`

函数就是把一段代码打包起来，想用的时候直接调用。

```python
def greet(name):
    print(f'Hello, {name}!')

greet('Wuyang')
greet('Python learner')
```
